**Q1.** Please answer the following questions in your own words.

1. How are trees constructed?

    Trees are built by repeatedly splitting the data based on feature values to create groups that are as pure as possible. Starting from a root node with all the data, a tree picks a feature and a threshold (for numerical variables) or category (for categorical variables) that best separates the data. This process continues recursively - creating branches and leaf nodes - until some stopping condition is met (like max depth, min number of samples in a node, or pure nodes).

2. How do trees handle non-linear relationships between variables? Compare this to linear models. 

     Trees are great for non-linear relationships because they don't assume a straight-line connection between features and outcomes. They split the space into regions using thresholds, which means they can model things like sharp changes or interactions easily. In contrast, linear models assume a straight-line (or linear plane) relationship between features and the outcome. That works well when the true relationship is linear, but it struggles with more complex patterns unless features are transformed or interactions are added manually.      
3. Why is the Gini a good loss function for categorical target variables? 
                                                                                       Gini impurity measures how often you would incorrectly label a randomly chosen element from a set if you assigned labels according to the distribution in a node. It's a solid choice because it's sensitive to how mixed the classes are in a node. The lower the Gini, the purer the node. That makes it a natural fit for classification problems, helping the tree decide the best splits.

4. Why do trees tend to overfit, and how can this tendency be constrained? 

     Trees overfit because they can keep splitting until they memorize the training data - especially when allowed to grow very deep. That makes them very flexible, but also prone to capturing noise. To prevent this, we can: limit the tree depth, set a minimum number of samples per leaf or per split, use pruning (removing unnecessary branches), use ensemble methods like Random Forests or Gradient Boosting, which reduce overfitting by averaging multiple trees.

5. True or false, and explain: Trees only really perform well in situations with lots of categorical variables as features/covariates. 

     False - Trees can handle both categorical and numerical variables quite well. They are often particularly convenient with categorical data (no need for dummy encoding), but they can also split numerical data based on thresholds. Their real strength is in handling complex patterns, interactions, and non-linearities - regardless of variable type.

6. Why don't most versions of classification/regression tree concept allow for more than two branches after a split?

     Most tree algorithms use binary splits (two branches) because it's computationally simpler and allows for more flexibility. Binary splits let the tree build more detailed decision boundaries step by step, which can often be more effective than making one big multi-way split. Also, evaluating all possible multi-way splits becomes very complex and slows down training significantly.

7. What are some heuristic ways you can examine a tree and decide whether it is probably over- or under-fitting?

     Signs of overfitting are the tree is very deep with many splits and leaves, it performs very well on training data but poorly on validation/test data, and leaf nodes have very few samples (high variance). Signs of underfitting are the tree is very shallow, it performs poorly on both training and test data, and itt doesn’t capture clear patterns in the data (high bias). One can also plot the tree and look at the splits - if they seem overly complex or too simple given what you know about the data, that can be a clue.

```python
#1

import pandas as pd

# Load the dataset
df = pd.read_csv(r"C:\Users\Brian\.vscode\trees\assignment\data\cars_hw.csv")

# Basic info
print(df.shape)         # Number of observations and features
print(df.columns)       # List available variables
print(df.head())        # First few rows
```

```python
#2

categorical_vars = ['Make', 'Color', 'Body_Type', 'No_of_Owners', 
                    'Fuel_Type', 'Transmission', 'Transmission_Type']
numeric_vars = ['Mileage_Run', 'Make_Year', 'Seating_Capacity']

# One-hot encode categorical variables
df_encoded = pd.get_dummies(df[categorical_vars + numeric_vars], drop_first=True)

# Add target variable
df_encoded['Price'] = df['Price']

#train-test split of data 
from sklearn.model_selection import train_test_split

X = df_encoded.drop(columns=['Price'])
y = df_encoded['Price']

#70% train, 30% test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

```

```python
#3
from sklearn.tree import DecisionTreeRegressor, plot_tree
from sklearn.metrics import r2_score

r2_scores = []
for leaf_size in range(1, 26):
    tree = DecisionTreeRegressor(min_samples_leaf=leaf_size, random_state=42)
    tree.fit(X_train, y_train)
    y_pred = tree.predict(X_test)
    r2_scores.append(r2_score(y_test, y_pred))

# Optimal value
best_leaf = range(1, 26)[r2_scores.index(max(r2_scores))]
print("Best min_samples_leaf:", best_leaf)
```

```python
#4

from sklearn.metrics import mean_squared_error
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Train with optimal leaf size
best_tree = DecisionTreeRegressor(min_samples_leaf=best_leaf, random_state=42)
best_tree.fit(X_train, y_train)

# Predictions and metrics
y_pred = best_tree.predict(X_test)
r2_best = r2_score(y_test, y_pred)
rmse_best = np.sqrt(mean_squared_error(y_test, y_pred))
print("R²:", r2_best)
print("RMSE:", rmse_best)

# Plot tree
plt.figure(figsize=(20, 10))
plot_tree(best_tree, feature_names=X.columns, filled=True, max_depth=3)
plt.show()

# Residuals
residuals = y_test - y_pred
sns.histplot(residuals, kde=True)
plt.title("Residuals Distribution")
plt.xlabel("Residual")
plt.show()

# Check symmetry
print("Mean Residual:", residuals.mean())
print("Skewness:", residuals.skew())
```

The density of residuals is relatively symmetric and centered around 0.

**Q4.** This is a case study about trees using bond rating data. This is a dataset about bond ratings for different companies, alongside a bunch of business statistics and other data. Companies often have multiple reviews at different dates. We want to predict the bond rating (AAA, AA, A, BBB, BB, B, ..., C, D). Do business fundamentals predict the company's rating?

1. Load the `./data/corporate_ratings.csv` dataset. How many observations and variables does it contain? Print out the first few rows of data.

2.  Plot a histogram of the `ratings` variable. It turns out that the gradations of AAA/AA/A and BBB/BB/B and so on make it hard to get good results with trees. Collapse all AAA/AA/A ratings into just A, and similarly for B and C.

3. Use all of the variables **except** Rating, Date, Name, Symbol, and Rating Agency Name. To include Sector, make a dummy/one-hot-encoded representation and include it in your features/covariates. Collect the relevant variables into a data matrix $X$. 

4. Do a train/test split of the data and use a decision tree classifier to predict the bond rating. Including a min_samples_leaf constraint can raise the accuracy and speed up computation time. Print a confusion matrix and the accuracy of your model. How well do you predict the different bond ratings?

5. If you include the rating agency as a feature/covariate/predictor variable, do the results change? How do you interpret this?

```python
#1

import pandas as pd

# Load the dataset
df = pd.read_csv(r"C:\Users\Brian\.vscode\trees\assignment\data\corporate_ratings.csv")

# Check dimensions and column names
print("Shape of dataset:", df.shape)
print("Columns:", df.columns.tolist())

# Preview the first few rows
print(df.head())
```

```python
#2

import matplotlib.pyplot as plt
import seaborn as sns

# Plot histogram of Rating before collapsing
plt.figure(figsize=(10,6))
sns.countplot(data=df, x='Rating', order=df['Rating'].value_counts().index)
plt.xticks(rotation=45)
plt.title("Original Bond Ratings Distribution")
plt.show()
```

```python
#3

# Collapse the Rating variable
def simplify_rating(rating):
    if rating in ['AAA', 'AA', 'A']:
        return 'A'
    elif rating in ['BBB', 'BB', 'B']:
        return 'B'
    elif rating in ['CCC', 'CC', 'C']:
        return 'C'
    else:
        return 'D'

df['Rating_Collapsed'] = df['Rating'].apply(simplify_rating)

# Plot histogram of simplified ratings
plt.figure(figsize=(6,4))
sns.countplot(data=df, x='Rating_Collapsed', order=['A', 'B', 'C', 'D'])
plt.title("Collapsed Bond Ratings Distribution")
plt.show()

# Drop irrelevant columns listed in question
X = df.drop(columns=['Rating', 'Rating_Collapsed', 'Date', 'Name', 'Symbol', 'Rating Agency Name'])

# One hot encode Sector
X = pd.get_dummies(X, columns=['Sector'], drop_first=True)

# Target variable
y = df['Rating_Collapsed']
```

```python
#4

from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix, accuracy_score, ConfusionMatrixDisplay

# Split the data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Train decision tree with min_samples_leaf
clf = DecisionTreeClassifier(min_samples_leaf=10, random_state=42)
clf.fit(X_train, y_train)

# Predict
y_pred = clf.predict(X_test)

# Evaluate
acc = accuracy_score(y_test, y_pred)
print("Accuracy:", acc)

# Confusion matrix
cm = confusion_matrix(y_test, y_pred, labels=['A', 'B', 'C', 'D'])
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['A', 'B', 'C', 'D'])
disp.plot()
plt.title("Confusion Matrix: Tree without Rating Agency")
plt.show()
```

```python
#6

# Add rating agency as feature
X_with_agency = df.drop(columns=['Rating', 'Rating_Collapsed', 'Date', 'Name', 'Symbol'])
X_with_agency = pd.get_dummies(X_with_agency, columns=['Sector', 'Rating Agency Name'], drop_first=True)

# Train/test split
X_train2, X_test2, y_train2, y_test2 = train_test_split(X_with_agency, y, test_size=0.3, random_state=42)

# Fit new tree
clf2 = DecisionTreeClassifier(min_samples_leaf=10, random_state=42)
clf2.fit(X_train2, y_train2)
y_pred2 = clf2.predict(X_test2)

# Accuracy and Confusion Matrix
acc2 = accuracy_score(y_test2, y_pred2)
print("Accuracy with Rating Agency:", acc2)

cm2 = confusion_matrix(y_test2, y_pred2, labels=['A', 'B', 'C', 'D'])
disp2 = ConfusionMatrixDisplay(confusion_matrix=cm2, display_labels=['A', 'B', 'C', 'D'])
disp2.plot()
plt.title("Confusion Matrix: Tree with Rating Agency")
plt.show()
```

Including the rating agency as a feature leads to a somewhat noticeable increase in prediction accuracy (0.706 to 0.742) suggesting that the agency itself may encode info that helps predicting bond rating. When revieing the confusion matrices, 'A' and 'B' ratings were correctly classified the most (with 'B' being classified the most overall) while 'C' and 'D' ratings were classified correctly the least (with 'D' having no correct classications).
